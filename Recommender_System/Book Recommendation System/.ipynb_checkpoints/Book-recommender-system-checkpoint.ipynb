### Popularity Based Recommender System

* We show the highest rating top 50 books (average rating)
* we add those book only which contain min. 250 rating/votes in the books

```python
rating_with_name = ratings.merge(books, on='ISBN')
rating_with_name.head(3)
```

```python
# book title -- groupby becoz same book have different ISBN number 
num_rating_df = rating_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
print(num_rating_df.head(2))
num_rating_df.rename(columns = {'Book-Rating':'num_ratings'},inplace = True)
num_rating_df.head()
```

```python
# book name with avg.rating
avg_rating_df = rating_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating_df.rename(columns = {'Book-Rating':'avg_rating'},inplace = True)
avg_rating_df.head()
```

```python
# now merge the num_rating_df and avg_rating_df on the bases Book-title

popularity_df = num_rating_df.merge(avg_rating_df, on = 'Book-Title')
popularity_df.head()
```

*********************************

```python
# top 50 books
popularity_df = popularity_df[popularity_df['num_ratings']>=250].sort_values('avg_rating', ascending = False).head(50)
popularity_df
```

```python
# now merge the data with book so we get proper detail of the book 
popularity_df = popularity_df.merge(books,on = 'Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_rating']]
popularity_df.head()
```

```python
popularity_df['Image-URL-M'][0]
```

### Collaborative Filtering Based Recommender System

* we consider only those user **Most Experience**
* user>200 and book > 50
* A user which give more than 200 book rating and book which contain min 50 rating

```python
rating_with_name.head(2)
```

```python
# groupby on the bases of the 'User-ID' 
a = rating_with_name.groupby('User-ID').count()['Book-Rating'] >200
book_lover = a[a].index
book_lover
```

```python
# filtering on the bases of user
filter_rating = rating_with_name[rating_with_name['User-ID'].isin(book_lover)]
filter_rating.head()
```

* Total 10 lacks entry but only 4 lacks74 thousand entry occurs

```python
filter_rating.shape
```

```python
# filtering on the books
b =filter_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_book = b[b].index
```

```python
final_rating = filter_rating[filter_rating['Book-Title'].isin(famous_book)]
```

```python
final_rating = final_rating.drop_duplicates()
final_rating.shape
```

```python
final_rating.head()
```

```python
# pivot tabel-- pivot table
pt = final_rating.pivot_table(index='Book-Title', columns = 'User-ID',values = 'Book-Rating')
pt
```

* In case most of the value are null so fill the null value 

```python
pt.fillna(0, inplace=True)
pt
```

* 706 books and 810 are trusted users

#### vector
* each book can be represent in form of 810 dimensional cordinate system
* **Simillar vector** --- calculate the distance b/w --**cosine_similarity** 

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
similarity_score.shape
#706 book ka euclidean distance nikala h 
```

* function jisme aap ke book ka name dege and in output it reflect back 5 simillar name of that book

```python
# first find out the index value of any book
np.where(pt.index =='4 Blondes')
```

```python
a = np.where(pt.index == '4 Blondes')[0][0]
a
```

```python
similarity_score[a]
```

* we must perform the sorting and pick out most/best score, we must enumerate --> distance ke sath sath iteam ka index bhi show hoga 
* Sorting --> sorting is done on the bases of index not on the bases of distance
* so we pass the key value to sort
* [1:6] index is starting from the 2 nd position because in ist element it is itself and upto 5 steps 

```python
similar_iteams = sorted(list(enumerate(similarity_score[a])),key = lambda x:x[1],reverse=True)[1:6]
similar_iteams
```

* Now pull out all the index and fetch the data that require

```python
list(books[books['Book-Title'] == pt.index[553]].drop_duplicates('Book-Title')['Book-Title'].values)
```

```python
def recommend(book_name):
    # index fetch
    index = np.where(pt.index == book_name)[0][0]
    similar_iteams = sorted(list(enumerate(similarity_score[index])),key = lambda x:x[1],reverse = True)[1:6]
    
    data = []
    for i in similar_iteams:
        iteam = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        iteam.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        iteam.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        iteam.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(iteam)
        
    return data
    
```

```python
recommend('Year of Wonders')
```
