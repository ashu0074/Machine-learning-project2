```python
import pandas as pd
```

```python
import os
```

```python
df = pd.DataFrame()
```

```python
def combineData(foldername):
    global df
    filepath =f'/home/arcgate/Downloads/Datasets_Healthy_Older_People/{foldername}'  # formating is used it has more data(file)
    for filename in os.listdir(filepath):
    #     print(filename)
        if(filename!='README.txt'):
            temp  =  pd.read_csv(os.path.join(filepath,filename),header=None, index_col=None)
    #         print(temp)
            df = pd.concat([df,temp])
```

```python
combineData('S1_Dataset')
combineData('S2_Dataset')
```
